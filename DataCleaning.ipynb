```python
from google.colab import files
 
 
uploaded = files.upload()
```

```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['olympics.csv']))
print(df)
```

```python
def read():
  return df

olympics = read()
olympics.head()
```

```python
olympics = olympics.rename(columns=olympics.iloc[0])
olympics
```

```python
olympics.drop(olympics.index[0], inplace=True)
olympics
```

```python
from numpy import NaN
olympics.rename(columns = {NaN :"Country",
                            "№ Summer" : "Summer_Olympics",
                           "01 !" : "Summer_Golds",
                           "02 !" : "Summer_Silvers",
                           "03 !" : "Summer_Bronzes",
                            "№ Winter" : "Winter_Olympics",
                             "01 !" : "Winter_Golds",
                           "02 !" : "Winter_Silvers",
                           "03 !" : "Winter_bronzes",
                           "№ Games" : "Total_Games",
                           "01 !" : "Total_Golds",
                           "02 !": "Total_Silvers",
                           "03 !": " Total_Bronzes",
                           "Combined total": "Combined total_medals"}, inplace = True)
```
